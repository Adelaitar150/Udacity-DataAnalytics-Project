```python
plt.figure(figsize=(18,9))
Goal_kings=EPL.groupby('Name')['Goals'].sum().sort_values(ascending = False).head(10)

Goal_kings.plot.bar(title="Top Goal Scorers")
plt.grid(color='lightgrey', linestyle='-', linewidth=0.7, alpha=0.2)
plt.ylabel('Number of Goals ', color = 'black', fontsize = '17')
plt.xlabel('Name of Players', color = 'black', fontsize = '17');
```

* Harry Kane mananged to beat Mohammed Salah to win the golden boot by a single goal. Bruno came in third and Bamford fourth which is an impressive stat for a striker from a newly promoted club.

```python
Playmakers=EPL.groupby('Name')['Assists'].sum().sort_values(ascending = False).head(10)

plt.figure(figsize=(16,8))
Playmakers.plot.bar(title="Top Playmakers", fontsize = '15')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.7, alpha=0.2)
plt.ylabel('Number of Assists', color = 'black', fontsize = '15')
plt.xlabel('Name of Players', color = 'black', fontsize = '15');
```

* Harry Kane also walked away with the player maker award managing to finish the season with 14 assists. Brilliant performance from the forward. The two players from the city of manchester Bruno Fernandes and Kevin De Bruyne came in second with twelve assists each.  

```python
top_passes_comp=EPL.groupby('Name')['Passes_Attempted'].sum().sort_values(ascending = False).head(10)

plt.figure(figsize=(16,8))
top_passes_comp.plot.bar(title="Top 10 players with with the highest completed passes")
plt.grid(color='lightgrey', linestyle='-', linewidth=0.7, alpha=0.2)                        
plt.ylabel('Number of Passes', color = 'black', fontsize = '15')
plt.xlabel('Name of Players', color = 'black', fontsize = '15');
```

* The two fullbacks from liverpool Andrew Robertson and Trent Alexander-Arnold shared the top two positions for the highest passes completed by a player.Rodri and Pierre Hojbjerg the to Defensive midfielders came in 3rd and 4th.

```python
plt.figure(figsize=(16,8))
sns.regplot(data = EPL, y='Yellow_Cards', x='Starts');
plt.xlabel('Number of starts', fontsize = 15)
plt.ylabel('Number of Cards', fontsize = 15);
```

* There seems to be a a strong relationship between the number of starts and number of yellow cards awarded.

```python
plt.figure(figsize=(16,8))
sns.regplot(data = EPL, x='Age', y='Starts');
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Number of Starts', fontsize = 15);
plt.title('Correlation between Number of starts and Age',  fontsize = 15)
```

* There seems to be a weak correlation between the players age and their number of starts 

```python
plt.figure(figsize=(16,8))
sns.regplot(data = EPL, x='Penalty_Goals', y='Penalty_Attempted');
plt.xlabel('Penalty Goals', fontsize = 15)
plt.ylabel('Number of Penalty Attempted', fontsize = 13);
plt.title('Correlation between Number of Penalty Attempted and Penalty Goals',  fontsize = 15)
```

* This Chart shows a positive correlation between the two variables.The higher the penalties attempted, the higher the number of goals scored.

```python
plt.figure(figsize=(16,8))
sns.boxplot(x='Club', y='Age', data=EPL.sort_values(by='Age'))
plt.yticks(fontsize=12)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('Age', size=14)
plt.title('Age Distribution Across Premier League Clubs',  fontsize = 15)
```

* The boxplot shows the detailed age range from youngest to oldest players in the primier league. Arsenal an example , has its youngest player being 18 years old and its oldest being 33 years old.Whereas its median age varies between 24 and 25 years.
