## Investigation Overview
My objectives for this investigation was to come out with the overall performance of players in that particular season as well as some distinct player attributes which included positions, age and natioanality.



## Dataset Overview
The dataset contains statistics of the English Premier League 2020/21 season.It has all the players that played in the EPL and their various statistics which includes minutes played, matches started, percentage of passes completed, goals scored and assists created.The total number of players are 531 with their corresponding 17 attributes.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
EPL=pd.read_csv('EPL2021.csv')
```

## (Golden Boot)
The golden boot award is given to to the player who manages to score the highest league goals at the end of the 38 games.
Harry Kane ,the Tottenham Hotspurs foward managed to beat Mohammed Salah to win the golden boot by a single goal.He scored 23 goals. Bruno came in third and Bamford fourth which is an impressive stat for a striker from a newly promoted club.
It is  interesting to find Ilkay Gundogan on the list.

```python
plt.figure(figsize=(18,9))
Goal_kings=EPL.groupby('Name')['Goals'].sum().sort_values(ascending = False).head(10)

Goal_kings.plot.bar(title="Top Goal Scorers")
plt.grid(color='lightgrey', linestyle='-', linewidth=0.7, alpha=0.2)
plt.ylabel('Number of Goals ', color = 'black', fontsize = '20')
plt.xlabel('Name of Players', color = 'black', fontsize = '20');
```

## (Playmaker)
Playmaking is all about being able to assist scoring teammates by provindg the final pass which lead to the goal.For the 2020/21 season, Harry Kane also walked away with the player maker award managing to finish the season with 14 assists. Brilliant performance from the forward. The two players from the city of manchester Bruno Fernandes and Kevin De Bruyne came in second with twelve assists each.

```python
Playmakers=EPL.groupby('Name')['Assists'].sum().sort_values(ascending = False).head(10)

plt.figure(figsize=(18,9))
Playmakers.plot.bar(title="Top Playmakers")
plt.grid(color='lightgrey', linestyle='-', linewidth=0.7, alpha=0.2)
plt.ylabel('Number of Assists', color = 'black', fontsize = '20')
plt.xlabel('Name of Players', color = 'black', fontsize = '20');
```

## (Number of Cards VS Match Starts)
Does playing time affect the number of cards given to a player? This was a question of interest.Insights showed that there is a positve correlation that exists between the two variables.Players who spent more time on the field were booked more.



```python
plt.figure(figsize=(16,8))
sns.regplot(data = EPL, y='Yellow_Cards', x='Starts');
plt.xlabel('Number of starts', fontsize = 20)
plt.ylabel('Number of Cards', fontsize = 20);
```

## (Average Club Age)
I found the age range of  the 20 premier league clubs worth investigating.With the aid of a boxplot, interesting insights were made.Manchester united has the youngest squad on average followed by southampton.Tottenham Hotspurs has the youngest player in the league with Chelsea having the oldest.



```python
plt.figure(figsize=(16,8))
sns.boxplot(x='Club', y='Age', data=EPL.sort_values(by='Age'))
plt.yticks(fontsize=12)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('Age', size=14)
```
